## Model evaluation in Regression Models

Accuracy metrics:
* training accuracy: percentage of correct prediction when using the test dataset.
  * A high percentage may indicate over-fitting -> trained to tight to the training data.
* out of sample accuracy: percentage of correct predictions based on data the model has not yet seen.
  * This is disired to transfer the model to new scenarios

> alternative names? training loss?

Possible approaches:
1. Train using the complete dataset and use a portion of the same dataset to test
 * Test set is part of the train set.
 * High training accuracy
 * Low out of sample accuracy
2. Train/Test split: split the dataset into two: one part of training and one part for testing (ofc wihout the labels)
 * Test set and train set are exclusive.
 * More accurate "out of sample accuracy"
 * Performance may depends on the split.

**k-fold cross-validation**
A method to determine the performance dependency on the splitting of data.

Using (all) variations on the data split to train and test the data. For example, a 4 fold would split the data into 25% for testing and the remaining 75% for training. This can be repeated 4 times by moving the test data selecting by a qaurter of the data each step. Note that the model is rebuild on each iteration. For each iteration, the accuracy is kept and finally averaged to indicate the effect on the input split on "out of sample accuracy".

**Error metrics**

Next step compare predicted values to the actual values (the residuals).
Error metrics: MSE
* MAE: mean absolute error (easy to interpret)
* MSE: mean square error (placed more weight on bigger errors)
* RMSE: root mean square error (same as MSE but the output can be interpretted in terms of the input (data dimensions))
* RAE: relative absolute error (relates the residual to a trivial (average) predictor)
* RSE: relative squared error
* R^2: popular metric for model accuracy, higher values indicate better fit.
